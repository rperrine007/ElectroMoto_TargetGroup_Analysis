## Importiere Bibliotheken

```python
import pandas as pd
```

## Importiere Bearbeitete Daten

```python
# Definiere Filepath
daten_path = "./data/df_gesamt.csv"
df_gesamt = pd.read_csv(daten_path)
```

## Erstelle Dataframe für Metrik

```python
# Zuweisung des Bewertungsschemas
metrik_bev_jung = 0.03 # Bevölkerung: Anteil 20-39-Jährige [%]
metrik_bev_alt = 0.07 # Bevölkerung: Anteil 40-64-Jährige [%]
metrik_motorrad_pro_einwohner = 0.03 # Motorräder pro 1000 Einwohner [Anz.]
metrik_motorrad_neuzulassungen = 0.04 # Motorrad-Neuzulasssungen pro 1000 Einw. [Anz.]
metrik_weg_haltestelle = 0.1 # Weg zur nächsten Haltestelle [m]
metrik_MV_wege = 0.2 # MIV-Wege Quell-, Ziel- und Binnenverkehr [Anz.]
metrik_erschliessung_öv = 0.1 # Erschliessung durch S-Bahn+Bus [% der Einw.] 
metrik_ladestationen_anz = 0.1 # electric_car_charging_spot_count
metrik_auto_per_ladestationen = 0.1 # electric_cars_per_charging_spot
metrik_neuzulassungen_elektromotor = 0.15 # Neuzulassungen  Elektromotor [%] 
metrik_elektromotor_bestand = 0.05 # Bestand Elektromotor [%]
metrik_steuerbares_einkommen = 0.03
```

```python
df_metrik = pd.DataFrame()

# Setze Eintrag 1 = Metrik und Eintrag 2 = Ranking-Method
df_metrik.at[1, 'Bevölkerung: Anteil 20-39-Jährige [%] '] = metrik_bev_jung
# Datentyp der Spalte zu 'object' ändern. So stellt es kein Problem mehr da dass im Dataframe verschiedene Datentypen vorhanden sind.
df_metrik['Bevölkerung: Anteil 20-39-Jährige [%] '] = df_metrik['Bevölkerung: Anteil 20-39-Jährige [%] '].astype(object)
# Ranking nach "min" oder "max" bewerten
df_metrik.at[2, 'Bevölkerung: Anteil 20-39-Jährige [%] '] = 'min'

df_metrik.at[1, 'Bevölkerung: Anteil 40-64-Jährige [%] '] = metrik_bev_alt
df_metrik['Bevölkerung: Anteil 40-64-Jährige [%] '] = df_metrik['Bevölkerung: Anteil 40-64-Jährige [%] '].astype(object)
df_metrik.at[2, 'Bevölkerung: Anteil 40-64-Jährige [%] '] = 'min'

df_metrik.at[1, 'Motorräder pro 1000 Einwohner [Anz.] '] = metrik_motorrad_pro_einwohner
df_metrik['Motorräder pro 1000 Einwohner [Anz.] '] = df_metrik['Motorräder pro 1000 Einwohner [Anz.] '].astype(object)
df_metrik.at[2, 'Motorräder pro 1000 Einwohner [Anz.] '] = 'min'

df_metrik.at[1, 'Motorrad-Neuzulasssungen pro 1000 Einw. [Anz.] '] = metrik_motorrad_neuzulassungen
df_metrik['Motorrad-Neuzulasssungen pro 1000 Einw. [Anz.] '] = df_metrik['Motorrad-Neuzulasssungen pro 1000 Einw. [Anz.] '].astype(object)
df_metrik.at[2, 'Motorrad-Neuzulasssungen pro 1000 Einw. [Anz.] '] = 'min'

df_metrik.at[1, 'Weg zur nächsten Haltestelle [m] '] = metrik_weg_haltestelle
df_metrik['Weg zur nächsten Haltestelle [m] '] = df_metrik['Weg zur nächsten Haltestelle [m] '].astype(object)
df_metrik.at[2, 'Weg zur nächsten Haltestelle [m] '] = 'min'

df_metrik.at[1, 'MIV-Wege Quell-, Ziel- und Binnenverkehr [Anz.] '] = metrik_MV_wege
df_metrik['MIV-Wege Quell-, Ziel- und Binnenverkehr [Anz.] '] = df_metrik['MIV-Wege Quell-, Ziel- und Binnenverkehr [Anz.] '].astype(object)
df_metrik.at[2, 'MIV-Wege Quell-, Ziel- und Binnenverkehr [Anz.] '] = 'min'

df_metrik.at[1, 'Erschliessung durch S-Bahn+Bus [% der Einw.] '] = metrik_erschliessung_öv
df_metrik['Erschliessung durch S-Bahn+Bus [% der Einw.] '] = df_metrik['Erschliessung durch S-Bahn+Bus [% der Einw.] '].astype(object)
df_metrik.at[2, 'Erschliessung durch S-Bahn+Bus [% der Einw.] '] = 'max'

df_metrik.at[1, 'electric_car_charging_spot_count'] = metrik_ladestationen_anz
df_metrik['electric_car_charging_spot_count'] = df_metrik['electric_car_charging_spot_count'].astype(object)
df_metrik.at[2, 'electric_car_charging_spot_count'] = 'min'

df_metrik.at[1, 'electric_cars_per_charging_spot' ] = metrik_auto_per_ladestationen
df_metrik['electric_cars_per_charging_spot'] = df_metrik['electric_cars_per_charging_spot' ].astype(object)
df_metrik.at[2, 'electric_cars_per_charging_spot' ] = 'max'

df_metrik.at[1, 'Neuzulassungen  Elektromotor [%] ' ] = metrik_neuzulassungen_elektromotor
df_metrik['Neuzulassungen  Elektromotor [%] '] = df_metrik['Neuzulassungen  Elektromotor [%] '].astype(object)
df_metrik.at[2, 'Neuzulassungen  Elektromotor [%] ' ] = 'min'

df_metrik.at[1, 'Bestand Elektromotor [%] ' ] = metrik_elektromotor_bestand
df_metrik['Bestand Elektromotor [%] ' ] = df_metrik['Bestand Elektromotor [%] ' ].astype(object)
df_metrik.at[2, 'Bestand Elektromotor [%] ' ] = 'min'

df_metrik.at[1, 'Ø steuerbares Einkommen natürliche Pers. [Fr.] ' ] = metrik_steuerbares_einkommen
df_metrik['Ø steuerbares Einkommen natürliche Pers. [Fr.] ' ] = df_metrik['Ø steuerbares Einkommen natürliche Pers. [Fr.] ' ].astype(object)
df_metrik.at[2, 'Ø steuerbares Einkommen natürliche Pers. [Fr.] ' ] = 'min'
```

## Bewerte Daten mit Rängen

```python
# finde gleiche Kolonnen
common_columns = df_gesamt.columns.intersection(df_metrik.columns)

# erstelle Dataframe
df_gesamt_ranked = pd.DataFrame()

# Sort values in common columns
for col in common_columns:
     # Methode aus df_metrik holen
    rank_method = df_metrik[col].iloc[1] 
    # Ascending für 'min', sonst 'max'
    ascending = True if rank_method == 'min' else False  
    # eigentliches ranking
    df_gesamt_ranked[col] = df_gesamt[col].rank(ascending=ascending, method='dense')
```

## Erstelle finale Bewertung

```python
# erstelle Dataframe
df_result = pd.DataFrame()

# Bewerte Ranking nach der erstellten Metrik
#Iteriere durch alle Kolonnen beider dataframes
for i in range(0, len(df_gesamt_ranked.columns)):
    for j in range(0, len(df_metrik.columns)):
        # Wenn Kolonnen-Namen gleich, rechne Metrik * Spalte der Ranking-Tabelle
        if(df_gesamt_ranked.columns[i] == df_metrik.columns[j]):
            column_name = df_gesamt_ranked.columns[i]
            df_result[column_name] = df_metrik.iloc[0, j] * df_gesamt_ranked[column_name][:]  
```

```python
# erstelle Dataframe
df_ranking = pd.DataFrame()

# Summe pro Zeile berechnen und als neue Spalte hinzufügen
df_result['Ranking Gesamt [Punkte]'] = df_result.sum(axis=1)

# bfs_nr und municipality wird für die spätere Identifizierung der Gemeinden benötigt. 
df_result['bfs_nr'] = df_gesamt['bfs_nr']
df_result['municipality'] = df_gesamt['municipality']

df_result.head()
```

```python
# Füge eine Rangliste anhand des erstellten Rankings hinzu.
df_result["Rangliste"] = df_result["Ranking Gesamt [Punkte]"].rank(ascending=False, method='dense')
```

## Ausgabe des resultierenden Files

```python
#Zwischenspeicherung
df_result.to_csv("./data/df_ranking.csv", index=False)
```
