## Importiere Bibliotheken

```python
import pandas as pd
```

## Importiere Bearbeitete Daten

Lade Gemeindedaten

```python
# Definiere Filepath
gmeindedaten_path = "./data/dfs_gemeindedaten_komplett.csv"
df_gemeindedaten = pd.read_csv(gmeindedaten_path)
```

Lade Energyreporter Daten

```python
energyreporter_path = "./data/dfs_energyreporter_municipality_cleansed.csv"
df_energyreporter = pd.read_csv(energyreporter_path)
```

## Bereite Daten für Zusammenfügen vor
Nehme nur Gemeindedaten mit Jahr = 2023 da Energy-Reporterdaten frisch 2024 gemessen wurden. Daher macht es nur Sinn die aktuellsten Daten zu kombinieren.

```python
#Filtere nach 2023 im Gemeindesatz-Daten
df_gemeindedaten_2023 = df_gemeindedaten[df_gemeindedaten["JAHR"] == 2023]
# Behalte Gebietname von energy-reporter Daten und lösche diesen. 
# Grund: Gebiet_Name in Gemeindedaten abgekürzt. # Für Erstellung der Heat-Map wird dies später Probleme verursachen.
df_gemeindedaten_2023 = df_gemeindedaten_2023.drop(columns=['GEBIET_NAME'])
# Gleiche Schreibweise BFS_NR den Energy-Reporter Daten an.
df_gemeindedaten_2023.rename(columns={'BFS_NR': 'bfs_nr'}, inplace=True)
```

```python
#lösche unnötige Spalten von Energyreporter
df_energyreporter_short = df_energyreporter.drop(columns=['electric_car_charging_spot_last_change'])
```

```python
# (ZH) aus der Spalte "municipality" entfernen
df_energyreporter_short['municipality'] = df_energyreporter_short['municipality'].str.replace(r' \(ZH\)$', '', regex=True)
```

## Füge Daten zusammen

```python
df_gesamt = pd.merge(df_gemeindedaten_2023, df_energyreporter_short, on=["bfs_nr"])
```

```python
# Kontrolle das keine weiteren NaN Werte existieren
df_gesamt[df_gesamt.isna().any(axis=1)].shape
```

```python
#Zwischenspeicherung
df_gesamt.to_csv("./data/df_gesamt.csv", index=False)
```
